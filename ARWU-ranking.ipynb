### Function

```python
def get_one_score(soup, build):
    length = 5 if build else 1
    result_list = [[] for i in range(length)]

    # for each row
    for item in soup.select("tr"):
        td = item.select("td")
        if(len(td) == 0):
            continue
        row = []

        try:
            if(build):
                # ranking
                row.append(td[0].find(class_="ranking").contents[0].lstrip().rstrip())  
                # university name
                try:
                    row.append(td[1].find(class_="univ-name").contents[0].lstrip().rstrip())
                except:
                    row.append(td[1].find(class_="univ-name-normal").contents[0].lstrip().rstrip())
                # regional rank
                row.append(td[3].contents[0].lstrip().rstrip())
                # total score
                row.append(td[4].contents[0].lstrip().rstrip())
            # item score
            row.append(td[5].contents[0].lstrip().rstrip())
        except:
            if(build):
                print("break error")
            else:
                print("item error")
            continue

        for index in range(len(result_list)):
            result_list[index].append(row[index])

    return result_list
```

```python
def write_to_csv(ranks, name, regional, total, scores):
    indicators = ["Alumni", "Award", "HiCi", "N&S", "PUB", "PCP"]

    df = pd.DataFrame()
    df["World Rank"] = ranks
    df["Institution"] = name
    df["National/Regional Rank"] = regional
    df["Total Score"] = total

    for index in range(len(indicators)):
        df[indicators[index]] = scores[index]

    # write file with append method
    df.to_csv("ARWU-ranking.csv", mode='a', header=False, index=False, encoding="utf-8")
    
```

```python
indicators = ["Alumni", "Award", "HiCi", "N&S", "PUB", "PCP"]
file = open("ARWU-ranking.csv", 'w')
file.write("World Rank,Institution,National/Regional Rank,Total Score,")
for index in range(len(indicators)):
    file.write(indicators[index] + ",")
file.write("\n")
file.close()

option = Options()
option.add_argument("--disable-notifications")
chrome = webdriver.Chrome('./chromedriver', options=option)

url = "https://www.shanghairanking.com/rankings/arwu/2020"
chrome.get(url)
soup = BeautifulSoup(chrome.page_source, 'html.parser')

page_count = 1
while(page_count <= 34):
    time.sleep(1)
    print("current page:", page_count)
    ranks, name, regional, total = [], [], [], []
    scores = [[] for i in range(6)]

    # for 6 item score
    btn_name_list = ["Alumni", "Award", "HiCi", "N&S", "PUB", "PCP"]

    for index in range(6):
        btns = chrome.find_elements_by_class_name("head-bg", )
        btns[-1].click()
        print(indicators[index], end=" | ")

        links = chrome.find_elements_by_tag_name("li")
        for link in links:
            if(link.text == btn_name_list[index]):
                link.click()
                soup = BeautifulSoup(chrome.page_source, 'lxml')

                if(index == 0):
                    result = get_one_score(soup, True)
                    ranks += result[0]
                    name += result[1]
                    regional += result[2]
                    total += result[3]
                    scores[index] += result[4]
                else:
                    result = get_one_score(soup, False)
                    scores[index] += result[0]

                time.sleep(2)
                break  # to click option button
    
    print()
    # click next page
    btns_next_page = chrome.find_elements_by_class_name("ant-pagination-item-link", )
    btns_next_page[-1].click()
    # scroll to top
    js="var action = document.documentElement.scrollTop = 0"
    chrome.execute_script(js)
    time.sleep(2)

    # write to file
    write_to_csv(ranks, name, regional, total, scores)

    page_count += 1

chrome.close()
print("finish")
```
